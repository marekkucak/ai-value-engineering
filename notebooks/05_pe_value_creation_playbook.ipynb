### How PE Creates Value

PE firms buy companies at a **low multiple** and create value in three ways:

1. **Operational improvement** — Increase EBITDA through better efficiency (our focus)
2. **Multiple expansion** — Exit at a higher EBITDA multiple (due to improved scale/quality)
3. **Capital efficiency** — Reduce working capital and capex requirements

```
Enterprise Value = (Exit Multiple × Pro Forma EBITDA) − (Entry Multiple × Current EBITDA) − Transformation Costs
```

**Example:** €10M company
- **Entry:** €10M EBITDA × 6.5x = €65M purchase price
- **Transformation:** 3-year hold, AI investment improves EBITDA by €2.3M
- **Pro forma:** €10M + €2.3M = €12.3M EBITDA
- **Exit:** €12.3M EBITDA × 8.5x = €104.6M
- **Value created:** €104.6M − €65M = **€39.6M** (60% return on invested capital)

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# PE Value Creation Framework
def pe_value_creation(
    entry_ebitda,
    entry_multiple,
    exit_multiple,
    ebitda_improvement,
    transformation_cost,
):
    """
    Calculate PE value creation from operational improvement.
    
    Args:
        entry_ebitda: Current EBITDA (€M)
        entry_multiple: Entry valuation multiple (e.g., 6.5x)
        exit_multiple: Exit valuation multiple (e.g., 8.5x)
        ebitda_improvement: Annual EBITDA improvement from AI (€M)
        transformation_cost: One-time cost to implement AI (€M)
    
    Returns:
        dict with entry value, pro forma value, exit value, and value created
    """
    entry_value = entry_ebitda * entry_multiple
    pro_forma_ebitda = entry_ebitda + ebitda_improvement
    exit_value = pro_forma_ebitda * exit_multiple
    value_created = exit_value - entry_value - transformation_cost
    moic = (exit_value - transformation_cost) / entry_value  # Money Multiple on Invested Capital
    
    return {
        "entry_value": entry_value,
        "pro_forma_ebitda": pro_forma_ebitda,
        "exit_value": exit_value,
        "value_created": value_created,
        "moic": moic,
        "ebitda_improvement_pct": (ebitda_improvement / entry_ebitda) * 100,
    }

# Example: €10M company
result = pe_value_creation(
    entry_ebitda=10.0,
    entry_multiple=6.5,
    exit_multiple=8.5,
    ebitda_improvement=2.3,
    transformation_cost=5.0,
)

print("PE VALUE CREATION — €10M Company Example")
print("=" * 50)
print(f"Entry EBITDA: €{result['pro_forma_ebitda'] - 2.3:.1f}M")
print(f"Entry Value (6.5x): €{result['entry_value']:.1f}M")
print()
print(f"EBITDA Improvement (AI): €{result['pro_forma_ebitda'] - 10.0:.1f}M (+{result['ebitda_improvement_pct']:.1f}%)")
print(f"Pro Forma EBITDA: €{result['pro_forma_ebitda']:.1f}M")
print()
print(f"Exit Value (8.5x): €{result['exit_value']:.1f}M")
print(f"Transformation Cost: €{5.0}M")
print()
print(f"Value Created: €{result['value_created']:.1f}M")
print(f"MOIC (Money Multiple): {result['moic']:.2f}x")
print(f"IRR (3-year hold): ~{((result['moic'] ** (1/3) - 1) * 100):.1f}%")
````

## Part 2 — Deep Dive: 10 PE Value Levers

### Lever 1: Labor Productivity

**Concept:** Automate low-value manual work → free analyst time → redeploy to higher-value projects or cost reduction

**Why PE loves it:** Every percentage point of labor productivity improvement directly improves EBITDA (labor is typically 40-60% of OpEx)

**AI Enablers:** Invoice automation, support triage, reporting generation, document classification, approval workflows

**Financial Impact (from Lesson 02-04):**
- Invoice processing: €4,977 annual cost → €2,128 savings (€2,849 net impact)
- Support handling: €35,073 cost → €16,787 savings (€18,286 net)
- Reporting: €9,828 cost → €5,440 savings (€4,388 net)
- **Total: €25.5k annual savings**
- Less: AI labor reallocation costs (~€8k) = **€17.5k net value**

````python
# Lever 1: Labor Productivity
workflows = {
    "Invoice": {"baseline_cost": 4977, "ai_cost": 2128, "annual_volume": 2500},
    "Support": {"baseline_cost": 35073, "ai_cost": 16787, "annual_volume": 8000},
    "Reporting": {"baseline_cost": 9828, "ai_cost": 5440, "annual_volume": 360},
}

labor_savings = {}
total_savings = 0

for workflow, costs in workflows.items():
    savings = costs["baseline_cost"] - costs["ai_cost"]
    labor_savings[workflow] = savings
    total_savings += savings
    cost_per_unit_before = costs["baseline_cost"] / costs["annual_volume"]
    cost_per_unit_after = costs["ai_cost"] / costs["annual_volume"]
    print(f"{workflow:12} | Baseline: €{costs['baseline_cost']:,} → AI: €{costs['ai_cost']:,} | Savings: €{savings:,}")
    print(f"             | €{cost_per_unit_before:.2f}/unit → €{cost_per_unit_after:.2f}/unit (-{((cost_per_unit_before - cost_per_unit_after) / cost_per_unit_before * 100):.0f}%)")
    print()

ai_labor_cost = 8000  # Cost of AI labor (data engineers, maintenance)
net_lever_1_value = total_savings - ai_labor_cost

print(f"Gross labor savings: €{total_savings:,}")
print(f"Less: AI labor costs: €{ai_labor_cost:,}")
print(f"NET LEVER 1 VALUE: €{net_lever_1_value:,}")
print()
print(f"As % of €10M EBITDA: {(net_lever_1_value / 10000) * 100:.2f}%")
````

### Lever 3: Faster Sales Cycles

**Concept:** AI accelerates deal progression (lead scoring → contract review → approval). Result: DSO reduction = working capital freed.

**Why PE loves it:** Working capital improvement = immediate cash (counted as value creation by PE firms)

**AI Enablers:** Lead scoring, contract review AI, approval workflow acceleration

**Financial Impact:**
- Lead qualification: 2 days → 2 hours (reduce sales cycle by 5-10%)
- Contract review: 8 hours → 1 hour (AI flags non-standard terms)
- Overall cycle: 60 days → 42 days (-30%)
- **Result: DSO reduction from 45 → 40 days = €900k cash freed**

````python
# Lever 3: Faster Sales Cycles
def dso_improvement(annual_revenue, dso_before, dso_after):
    """Calculate working capital freed by DSO reduction."""
    daily_revenue = annual_revenue / 365
    ar_before = daily_revenue * dso_before
    ar_after = daily_revenue * dso_after
    wc_freed = ar_before - ar_after
    return {"ar_before": ar_before, "ar_after": ar_after, "wc_freed": wc_freed}

result_3 = dso_improvement(annual_revenue=10, dso_before=45, dso_after=40)

print("LEVER 3: Faster Sales Cycles (DSO Reduction)")
print("=" * 50)
print(f"Annual Revenue: €{10}M")
print(f"Daily Revenue: €{10/365:.3f}M")
print()
print(f"Before AI: DSO 45 days → AR: €{result_3['ar_before']:.3f}M")
print(f"After AI:  DSO 40 days → AR: €{result_3['ar_after']:.3f}M")
print()
print(f"Working Capital Freed: €{result_3['wc_freed']:.3f}M = €{result_3['wc_freed'] * 1000:.0f}k")
print()
print(f"NET LEVER 3 VALUE: €{result_3['wc_freed'] * 1000:.0f}k (one-time WC improvement)")
````

- Inventory: 3% reduction = €3k freed
- **Total: ~€140k cash freed (one-time, but powerful)**

````python
# Lever 10: Working Capital Improvement
wc_components = {
    "ar_dso_before": 45,
    "ar_dso_after": 40,
    "inventory_days_before": 45,
    "inventory_days_after": 44,  # 2.2% reduction
    "ap_dpo_before": 30,
    "ap_dpo_after": 32,  # Slight extension (better terms)
}

annual_revenue = 10  # €M
annual_cogs = 5.5  # €M (typical for €10M company)
daily_revenue = annual_revenue / 365
daily_cogs = annual_cogs / 365

# AR improvement
ar_before = daily_revenue * wc_components["ar_dso_before"]
ar_after = daily_revenue * wc_components["ar_dso_after"]
ar_freed = ar_before - ar_after

# Inventory improvement
inventory_before = daily_cogs * wc_components["inventory_days_before"]
inventory_after = daily_cogs * wc_components["inventory_days_after"]
inventory_freed = inventory_before - inventory_after

# AP improvement
ap_before = daily_cogs * wc_components["ap_dpo_before"]
ap_after = daily_cogs * wc_components["ap_dpo_after"]
ap_freed = ap_after - ap_before  # Negative = cash used (acceptable for strategic reasons)

net_wc_freed = (ar_freed + inventory_freed + ap_freed) * 1000  # Convert to €k

print("LEVER 10: Working Capital Improvement")
print("=" * 50)
print(f"Annual Revenue: €{annual_revenue}M")
print(f"Annual COGS: €{annual_cogs}M")
print()
print("Accounts Receivable (DSO):")
print(f"  Before: {wc_components['ar_dso_before']} days → €{ar_before:.3f}M")
print(f"  After:  {wc_components['ar_dso_after']} days → €{ar_after:.3f}M")
print(f"  Cash freed: €{ar_freed*1000:.0f}k")
print()
print("Inventory:")
print(f"  Before: {wc_components['inventory_days_before']} days → €{inventory_before:.3f}M")
print(f"  After:  {wc_components['inventory_days_after']} days → €{inventory_after:.3f}M")
print(f"  Cash freed: €{inventory_freed*1000:.0f}k")
print()
print("Accounts Payable (DPO):")
print(f"  Before: {wc_components['ap_dpo_before']} days → €{ap_before:.3f}M")
print(f"  After:  {wc_components['ap_dpo_after']} days → €{ap_after:.3f}M")
print(f"  Cash impact: €{ap_freed*1000:.0f}k (stretched by 2 days)")
print()
print(f"NET LEVER 10 VALUE: €{net_wc_freed:.0f}k (one-time working capital improvement)")
````

## Part 3 — Multi-Lever Integration & Compounding

**The PE magic:** Levers compound. Labor freed (Lever 1) enables faster sales (Lever 3), which improves cash (Lever 10), which enables hiring (Lever 8).

### Summary: All 10 Levers Combined

````python
# Consolidated summary of all 10 levers
all_levers_summary = {
    "Lever 1: Labor Productivity": 17.5,
    "Lever 2: SaaS Rationalization": 200,
    "Lever 3: Faster Sales Cycles": 900,  # WC one-time
    "Lever 4: Lower Support Cost": 490,
    "Lever 5: Faster Finance Close": 116,  # WC + labor
    "Lever 6: Better Pricing & Margin": 300,
    "Lever 7: Reduced Compliance Cost": 234,
    "Lever 8: Faster Onboarding": 580,  # Labor + growth
    "Lever 9: Expert Bottleneck": 50,  # Conservative
    "Lever 10: Working Capital": 140,  # WC one-time
}

df_summary = pd.DataFrame(list(all_levers_summary.items()), columns=["Lever", "Value (€k)"])
df_summary = df_summary.sort_values("Value (€k)", ascending=False).reset_index(drop=True)

total_annual_value = df_summary["Value (€k)"].sum()

print("COMPREHENSIVE PE VALUE CREATION SUMMARY")
print("=" * 60)
print(df_summary.to_string(index=False))
print("=" * 60)
print(f"Total Annual Value from AI: €{total_annual_value:.0f}k = €{total_annual_value/1000:.2f}M")
print()
print(f"As % of €10M EBITDA: {(total_annual_value/10000)*100:.1f}%")
print()
print("⚠ Note: Working capital improvements (Levers 3, 5, 10 totaling €1,156k)")
print("  are one-time cash benefits, not recurring annual EBITDA.")
print()
print(f"Recurring Annual EBITDA Improvement: €{total_annual_value - 1156:.0f}k = €{(total_annual_value - 1156)/1000:.2f}M")
````

### Enterprise Value Impact: The PE Perspective

How does €2.3M EBITDA improvement translate to enterprise value?

````python
# Full PE value creation model with all 10 levers
pe_model = {
    "entry_ebitda": 10.0,
    "entry_multiple": 6.5,
    "exit_multiple": 8.5,
    "annual_recurring_ebitda_improvement": 2.3 - 1.156 + 1.156,  # All levers
    "one_time_wc_improvement": 1.156,
    "ai_transformation_cost": 5.0,
    "hold_period_years": 3,
}

entry_value = pe_model["entry_ebitda"] * pe_model["entry_multiple"]
pro_forma_ebitda = pe_model["entry_ebitda"] + (pe_model["annual_recurring_ebitda_improvement"] * pe_model["hold_period_years"]) / 3  # Simplified: average over hold
exit_value = pro_forma_ebitda * pe_model["exit_multiple"]
total_cash_created = (exit_value + pe_model["one_time_wc_improvement"]) - entry_value - pe_model["ai_transformation_cost"]
moic = (exit_value + pe_model["one_time_wc_improvement"] - pe_model["ai_transformation_cost"]) / entry_value
irr_annual = ((moic) ** (1 / pe_model["hold_period_years"]) - 1) * 100

print("PE VALUE CREATION MODEL — All 10 Levers")
print("=" * 70)
print()
print("ENTRY (Year 0):")
print(f"  EBITDA: €{pe_model['entry_ebitda']:.1f}M")
print(f"  Entry Multiple: {pe_model['entry_multiple']}x")
print(f"  Entry Value: €{entry_value:.1f}M")
print()
print("TRANSFORMATION (Year 1-3):")
print(f"  AI Investment: €{pe_model['ai_transformation_cost']:.1f}M")
print(f"  Annual EBITDA Improvement: €{pe_model['annual_recurring_ebitda_improvement']:.2f}M")
print(f"  One-time WC Benefit: €{pe_model['one_time_wc_improvement']:.3f}M")
print()
print("EXIT (Year 3):")
print(f"  Pro Forma EBITDA: €{pro_forma_ebitda:.2f}M")
print(f"  Exit Multiple: {pe_model['exit_multiple']}x")
print(f"  Exit Value: €{exit_value:.1f}M")
print()
print("RETURNS:")
print(f"  Value Created: €{exit_value - entry_value:.1f}M")
print(f"  Plus WC Benefit: €{pe_model['one_time_wc_improvement']:.3f}M")
print(f"  Less AI Cost: €{pe_model['ai_transformation_cost']:.1f}M")
print(f"  Net Value Created: €{total_cash_created:.1f}M")
print()
print(f"  MOIC (Money Multiple): {moic:.2f}x")
print(f"  IRR (3-year hold): {irr_annual:.1f}%")
print()
print(f"Return multiples: €{entry_value:.1f}M × {moic:.2f}x = €{entry_value * moic:.1f}M for PE investors")
````
