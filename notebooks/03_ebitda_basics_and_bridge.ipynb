## Part 1: EBITDA Fundamentals

### What is EBITDA?
**EBITDA** = Earnings Before Interest, Taxes, Depreciation, Amortization

In simpler terms:
```
EBITDA = Revenue − Cost of Goods Sold − Operating Expenses
```

### Why do PE firms care about EBITDA?
- It's **cash-like profit** — close to the actual cash flow available to equity owners
- It removes accounting artifacts (depreciation, interest) that mask real cash generation
- It's the foundation for valuation multiples (typically 6x-12x EBITDA in M&A)

### EBITDA Margins by Industry
- **SaaS:** 40-60% (high gross margins, scalable)
- **Professional Services:** 15-25% (people-based, lower leverage)
- **Manufacturing:** 10-15% (capital-intensive, commodity-like)
- **Our model company:** 15% (knowledge work, service-heavy)

## Part 2: EBITDA Bridge Building Blocks

### The EBITDA Bridge Concept
An **EBITDA bridge** shows how a company moves from current profitability to pro forma (forecast) profitability by stacking value levers.

### Six Bridge Components
1. **Current EBITDA** — Baseline profitability before AI
2. **+ Labor Savings** — Freed headcount or efficiency gains (Lesson 02)
3. **+ SaaS Savings** — License/subscription consolidation (Lesson 10)
4. **+ Revenue Uplift** — Margin improvement or new revenue (advanced)
5. **− AI Platform Cost** — Recurring platform, API, model hosting costs
6. **− Support/Maintenance** — Ongoing ops, retraining, monitoring

### Key Rule: No Double-Counting
Each component flows to EBITDA exactly once. Labor savings are counted as OpEx reduction (line item change), not as headcount cut savings AND efficiency gains.

````python
def ebitda_bridge(current_ebitda, labor_savings=0, saas_savings=0, 
                  revenue_uplift=0, ai_platform_cost=0, maintenance_cost=0,
                  quality_risk_buffer=0):
    """
    Calculate pro forma EBITDA using the EBITDA bridge.
    
    Args:
        current_ebitda: baseline EBITDA before AI (€)
        labor_savings: cost reduction from workflow efficiency (€/year)
        saas_savings: software cost reduction (€/year)
        revenue_uplift: additional revenue × margin (€/year)
        ai_platform_cost: recurring AI platform fee (€/year)
        maintenance_cost: support, retraining, monitoring (€/year)
        quality_risk_buffer: conservative discount for model risk (€/year)
    
    Returns:
        dict with bridge components and pro forma EBITDA
    """
    pro_forma_ebitda = (
        current_ebitda
        + labor_savings
        + saas_savings
        + revenue_uplift
        - ai_platform_cost
        - maintenance_cost
        - quality_risk_buffer
    )
    
    return {
        'current_ebitda': current_ebitda,
        'labor_savings': labor_savings,
        'saas_savings': saas_savings,
        'revenue_uplift': revenue_uplift,
        'ai_platform_cost': -ai_platform_cost,
        'maintenance_cost': -maintenance_cost,
        'quality_risk_buffer': -quality_risk_buffer,
        'pro_forma_ebitda': pro_forma_ebitda,
        'ebitda_delta': pro_forma_ebitda - current_ebitda,
        'ebitda_delta_pct': (pro_forma_ebitda - current_ebitda) / current_ebitda * 100 if current_ebitda > 0 else 0,
    }

print("✓ EBITDA bridge function defined")
````

## Part 4: Workflow 1 — Invoice Approval EBITDA Bridge

````python
# Invoice Approval EBITDA bridge
print("WORKFLOW 1: INVOICE APPROVAL EBITDA BRIDGE")
print("="*70)

inv = workflows_lesson02['Invoice Approval']

# Bridge components
inv_support_cost = 200  # Annual support (1 person × 5% time)
inv_quality_risk = 100  # Conservative risk buffer for AI errors

inv_bridge = ebitda_bridge(
    current_ebitda=company['current_ebitda'],
    labor_savings=inv['ai_savings'],
    ai_platform_cost=inv['ai_cost'],
    maintenance_cost=inv_support_cost,
    quality_risk_buffer=inv_quality_risk,
)

# Display
print(f"\nCurrent EBITDA:              €{inv_bridge['current_ebitda']:>12,.0f}")
print(f"+ Labor savings:             €{inv_bridge['labor_savings']:>12,.0f}  (AI efficiency)")
print(f"+ SaaS savings:              €{inv_bridge['saas_savings']:>12,.0f}  (none)")
print(f"+ Revenue uplift:            €{inv_bridge['revenue_uplift']:>12,.0f}  (none modeled)")
print(f"- AI platform cost:          €{inv_bridge['ai_platform_cost']:>12,.0f}")
print(f"- Support/maintenance:       €{inv_bridge['maintenance_cost']:>12,.0f}")
print(f"- Quality risk buffer:       €{inv_bridge['quality_risk_buffer']:>12,.0f}")
print("-" * 70)
print(f"Pro forma EBITDA:            €{inv_bridge['pro_forma_ebitda']:>12,.0f}")
print(f"\nEBITDA delta:                €{inv_bridge['ebitda_delta']:>12,.0f}  ({inv_bridge['ebitda_delta_pct']:>6.2f}%)")
print("\n→ Small but real EBITDA improvement: +0.10%")
````

## Part 7: Consolidated EBITDA Bridge (All 3 Workflows)

````python
# Consolidated bridge: stack all 3 workflows
print("CONSOLIDATED EBITDA BRIDGE: All 3 Workflows")
print("="*70)

# Total components
total_labor_savings = (inv['ai_savings'] + 
                       int(sup['ai_savings'] * 0.70) +  # Conservative scenario
                       rep['ai_savings'])

total_ai_costs = inv['ai_cost'] + sup['ai_cost'] + rep['ai_cost']
total_support = inv_support_cost + sup_support_cost + rep_support_cost
total_quality_risk = inv_quality_risk + sup_quality_risk + rep_quality_risk

consolidated_bridge = ebitda_bridge(
    current_ebitda=company['current_ebitda'],
    labor_savings=total_labor_savings,
    ai_platform_cost=total_ai_costs,
    maintenance_cost=total_support,
    quality_risk_buffer=total_quality_risk,
)

print(f"\nCurrent EBITDA:              €{consolidated_bridge['current_ebitda']:>12,.0f}")
print(f"\n+ LABOR SAVINGS:")
print(f"  Invoice Approval:          €{inv['ai_savings']:>12,.0f}")
print(f"  Support Triage (70%):      €{int(sup['ai_savings'] * 0.70):>12,.0f}")
print(f"  Reporting:                 €{rep['ai_savings']:>12,.0f}")
print(f"  Subtotal:                  €{consolidated_bridge['labor_savings']:>12,.0f}")

print(f"\n- AI & SUPPORT COSTS:")
print(f"  AI platform:               €{consolidated_bridge['ai_platform_cost']:>12,.0f}")
print(f"  Support/maintenance:       €{consolidated_bridge['maintenance_cost']:>12,.0f}")
print(f"  Quality risk buffer:       €{consolidated_bridge['quality_risk_buffer']:>12,.0f}")
print("-" * 70)
print(f"Pro forma EBITDA:            €{consolidated_bridge['pro_forma_ebitda']:>12,.0f}")
print(f"\nTOTAL EBITDA DELTA:          €{consolidated_bridge['ebitda_delta']:>12,.0f}")
print(f"EBITDA IMPROVEMENT:          {consolidated_bridge['ebitda_delta_pct']:>12.2f}%")
print("\n" + "="*70)
print("KEY INSIGHT:")
print(f"€{consolidated_bridge['ebitda_delta']:,.0f}/year in additional EBITDA from optimizing 3 workflows.")
print(f"That's {consolidated_bridge['ebitda_delta_pct']:.2f}% EBITDA growth without revenue increase!")
````
