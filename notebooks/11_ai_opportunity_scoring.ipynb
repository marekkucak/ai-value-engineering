> **Important:** Score = prioritisation signal, not a guarantee. Always validate
> the top-scored opportunities with a business owner before committing implementation resources.

## Part 1 — The AI Opportunity Scoring Framework

### The Formula

```
AI Opportunity Score =
    (value × frequency × reviewability × data_availability)
    ÷
    (risk × integration_complexity × change_resistance)
```

### Scoring Dimensions (each rated 1–10)

**Numerator — why this opportunity is attractive:**

| Dimension | What it measures | High (9–10) | Low (1–3) |
|-----------|----------------|-------------|-----------|
| value | EUR impact if implemented | >EUR 50k saving | <EUR 5k saving |
| frequency | How often the workflow runs | Daily, high volume | Monthly, low volume |
| reviewability | How easy for human to review AI output | Binary check, clear criteria | Complex judgment required |
| data_availability | Is structured input data available? | Clean system data, APIs exist | Manual spreadsheets, PDFs |

**Denominator — why this opportunity is difficult:**

| Dimension | What it measures | High (9–10) | Low (1–3) |
|-----------|----------------|-------------|-----------|
| risk | What goes wrong if AI makes an error | Financial/legal/regulatory harm | Low-stakes internal task |
| integration_complexity | How hard to connect to existing systems | Custom legacy ERP, no API | SaaS with REST API |
| change_resistance | How much pushback expected from users | Union, senior exec, regulated role | Tech-savvy team, low headcount impact |

### Connection to L06–L10

| Previous lesson | What it provides for scoring |
|----------------|------------------------------|
| L06 Cost of Friction | `value` (EUR saving), `frequency` (annual volume) |
| L07 Value Stream | `frequency` (step volume), `reviewability` (output clarity) |
| L08 Graph Bottlenecks | `integration_complexity` (system centrality), `risk` (bottleneck risk) |
| L09 Working Capital | `value` (cash release EUR), `data_availability` (ERP data quality) |
| L10 SaaS Rationalization | `integration_complexity` (bridge cost), `change_resistance` (tool adoption) |

## Part 2 — Scoring Utility Functions

These functions calculate opportunity scores and translate them into
prioritisation signals and payback metrics.

````python
def ai_opportunity_score(value, frequency, reviewability, data_avail, risk, integ_complex, change_resist):
    '''Calculate AI opportunity score using 7-dimension formula. Each dimension rated 1-10.'''
    numerator = value * frequency * reviewability * data_avail
    denominator = risk * integ_complex * change_resist
    if denominator == 0:
        return 0
    return numerator / denominator

def payback_months(impl_cost_eur, annual_saving_eur):
    '''Calculate payback period in months.'''
    if annual_saving_eur <= 0:
        return float('inf')
    return (impl_cost_eur / annual_saving_eur) * 12

def priority_label(score):
    '''Assign priority label based on score.'''
    if score >= 25:
        return 'HIGH'
    elif score >= 12:
        return 'MEDIUM'
    elif score >= 5:
        return 'LOW'
    else:
        return 'DEFER'

# Test functions
test_score = ai_opportunity_score(9, 9, 8, 8, 4, 5, 4)
test_payback = payback_months(12000, 28000)
print(f'Test score: {test_score:.1f}')
print(f'Test payback: {test_payback:.1f} months')
````

## Part 3 — Aggregated Opportunity Inventory

### The Master Opportunity List

We now aggregate all opportunities discovered across L06–L10 into one ranked pipeline.
Each opportunity includes:

- **Source lesson:** where it was originally analysed (friction cost, value stream, dependency graph, etc.)
- **Seven scoring dimensions:** all rated 1–10 based on L06–L10 evidence
- **Annual saving:** EUR impact if implemented
- **Implementation cost:** one-time project cost to deploy

Same portfolio company context as L06–L10: 200-person PE-backed professional services firm.

````python
opportunities = pd.DataFrame({
    'workflow': ['invoice_triage', 'support_summary', 'mgmt_reporting_automation',
        'lead_qualification', 'onboarding_doc_review', 'contract_extraction',
        'data_reconciliation', 'compliance_review', 'access_review_automation',
        'ar_aging_alerts', 'cash_flow_forecasting', 'dso_prediction',
        'saas_usage_monitoring', 'duplicate_tool_detection', 'bridge_automation_crm'],
    'source_lesson': ['L06', 'L06', 'L06', 'L07', 'L07', 'L07', 'L08', 'L08', 'L08',
        'L09', 'L09', 'L09', 'L10', 'L10', 'L10'],
    'value': [9, 8, 7, 8, 6, 7, 7, 6, 7, 9, 8, 7, 6, 7, 6],
    'frequency': [9, 10, 6, 9, 7, 6, 8, 5, 6, 9, 7, 6, 9, 5, 9],
    'reviewability': [8, 9, 7, 7, 8, 8, 7, 6, 8, 9, 7, 7, 9, 9, 7],
    'data_availability': [8, 7, 8, 7, 6, 6, 7, 6, 9, 9, 8, 7, 9, 8, 7],
    'risk': [4, 5, 4, 3, 4, 5, 6, 8, 7, 3, 5, 4, 2, 2, 3],
    'integration_complexity': [5, 4, 5, 4, 3, 4, 6, 5, 6, 3, 4, 4, 2, 2, 5],
    'change_resistance': [4, 3, 5, 5, 4, 4, 4, 6, 5, 3, 4, 3, 2, 4, 5],
    'annual_saving_eur': [28000, 38400, 16100, 22000, 12000, 18000, 20000, 15000, 16000,
        45000, 30000, 18000, 12000, 24000, 14000],
    'impl_cost_eur': [12000, 8000, 10000, 9000, 6000, 11000, 14000, 20000, 15000,
        5000, 12000, 8000, 3000, 4000, 8000],
})

opportunities['score'] = opportunities.apply(
    lambda r: ai_opportunity_score(r['value'], r['frequency'], r['reviewability'],
        r['data_availability'], r['risk'], r['integration_complexity'], r['change_resistance']),
    axis=1,
).round(1)
opportunities['payback_months'] = opportunities.apply(
    lambda r: payback_months(r['impl_cost_eur'], r['annual_saving_eur']),
    axis=1,
).round(1)
opportunities['priority'] = opportunities['score'].apply(priority_label)
opportunities = opportunities.sort_values('score', ascending=False).reset_index(drop=True)

total_opportunities = len(opportunities)
total_saving_potential = opportunities['annual_saving_eur'].sum()
total_impl_cost = opportunities['impl_cost_eur'].sum()

print('AGGREGATED AI OPPORTUNITY INVENTORY')
print('=' * 80)
print(f'Total opportunities: {total_opportunities}')
print(f'Total potential saving: EUR {total_saving_potential:,.0f}/year')
print(f'Total implementation cost: EUR {total_impl_cost:,.0f}')
print(f'Avg payback: {opportunities["payback_months"].mean():.1f} months')
print()
print('TOP 10 OPPORTUNITIES BY SCORE:')
print(opportunities[['workflow', 'source_lesson', 'score', 'priority', 'annual_saving_eur', 'payback_months']].head(10).to_string(index=False))
````

## Part 4 — Dimension Calibration Guide

### How to Rate Each Dimension Consistently

Consistent scoring requires anchoring. Use the table below to rate each opportunity
on each dimension. The key is to use the **same evidence threshold** across all opportunities.

#### Value (1–10): EUR annual saving if implemented

| Rating | Annual Saving | Examples |
|--------|---------------|----------|
| 9–10 | >EUR 40k | Eliminate 5+ FTE friction (L06); Release >20 days cash; Consolidate 3+ SaaS |
| 7–8 | EUR 20k–40k | Eliminate 2–3 FTE friction; Release 10–20 days cash |
| 5–6 | EUR 10k–20k | Eliminate 1 FTE friction; Release 5–10 days cash |
| 3–4 | EUR 3k–10k | Minor friction reduction; Small SaaS saving |
| 1–2 | <EUR 3k | Negligible impact; nice-to-have |

#### Frequency (1–10): How often the workflow runs

| Rating | Frequency | Examples |
|--------|-----------|----------|
| 9–10 | >5/day | Invoice triage, support tickets, AR aging |
| 7–8 | Daily | Lead qualification, data reconciliation |
| 5–6 | Weekly | Mgmt reporting, contract extraction, access review |
| 3–4 | Monthly | Consolidation reviews, cash forecasting |
| 1–2 | <Monthly | Annual planning, one-off migrations |

#### Reviewability (1–10): How easy to review AI output?

| Rating | Criteria | Examples |
|--------|----------|----------|
| 9–10 | Binary check (yes/no), clear audit trail | AR aging flags, SaaS deprecation alerts |
| 7–8 | Spot-check 10%, clear approval workflow | Invoice routing, support triage |
| 5–6 | Domain expert review required | Contract extraction, compliance flags |
| 3–4 | Complex judgment, multiple reviewers | Data reconciliation, forecasting models |
| 1–2 | Cannot easily verify correctness | Novel generative content |

#### Data Availability (1–10): Is structured data available?

| Rating | Data Source | Examples |
|--------|------------|----------|
| 9–10 | Clean ERP export, REST API, SSO logs | Workday, NetSuite, Datadog APIs |
| 7–8 | Mostly clean, occasional gaps | Email headers, PDF invoices with OCR |
| 5–6 | Mixed structured/unstructured | Email + attachments, Jira + notes |
| 3–4 | Mostly manual, sparse exports | Spreadsheets, manual process logs |
| 1–2 | Unstructured, requires heavy ETL | Handwritten notes, video call transcripts |

````python
top_opp = opportunities.iloc[0]
base_score = top_opp['score']

dimensions = ['value', 'frequency', 'reviewability', 'data_availability', 'risk', 'integration_complexity', 'change_resistance']
sensitivities = []

for dim in dimensions:
    test_vals_up = top_opp.to_dict()
    test_vals_up[dim] = min(10, test_vals_up[dim] + 2)
    score_up = ai_opportunity_score(test_vals_up['value'], test_vals_up['frequency'],
        test_vals_up['reviewability'], test_vals_up['data_availability'],
        test_vals_up['risk'], test_vals_up['integration_complexity'], test_vals_up['change_resistance'])
    delta_up = score_up - base_score
    
    test_vals_dn = top_opp.to_dict()
    test_vals_dn[dim] = max(1, test_vals_dn[dim] - 2)
    score_dn = ai_opportunity_score(test_vals_dn['value'], test_vals_dn['frequency'],
        test_vals_dn['reviewability'], test_vals_dn['data_availability'],
        test_vals_dn['risk'], test_vals_dn['integration_complexity'], test_vals_dn['change_resistance'])
    delta_dn = score_dn - base_score
    sensitivities.append({'dimension': dim, 'delta_up': delta_up, 'delta_dn': delta_dn})

sens_df = pd.DataFrame(sensitivities).sort_values('delta_up', ascending=True)

fig, ax = plt.subplots(figsize=(10, 5))
x_pos = range(len(sens_df))
ax.barh(x_pos, sens_df['delta_up'], color='#4dac26', label='Improve +2', alpha=0.8)
ax.barh(x_pos, sens_df['delta_dn'], color='#d73027', label='Worsen -2', alpha=0.8)
ax.set_yticks(x_pos)
ax.set_yticklabels(sens_df['dimension'])
ax.set_xlabel('Score Delta')
ax.set_title(f'Sensitivity Analysis: {top_opp["workflow"]} (base score: {base_score:.1f})')
ax.legend()
ax.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.savefig('sensitivity_tornado.png', dpi=100, bbox_inches='tight')
plt.show()
````

## Part 5 — The Priority Quadrant Framework

### Four Quadrant Matrix

**Quadrant I — Quick Wins** (high score, low effort, fast payback)
- Opportunity score ≥ 20 AND implementation effort ≤ EUR 10k AND payback ≤ 12 months
- Action: Start in 100-day Wave 1 (Day 1–30)

**Quadrant II — Strategic Projects** (high score, high effort, justified by scale)
- Opportunity score ≥ 20 AND implementation effort > EUR 10k
- Action: Start in Wave 2 or Wave 3 after quick wins establish momentum

**Quadrant III — Defer or Skip** (low score, high effort, poor ROI)
- Opportunity score < 12 OR payback > 18 months
- Action: Revisit only if priorities shift or input assumptions change

**Quadrant IV — Fill-Ins** (low score, low effort, good for team capacity)
- Opportunity score 5–12 AND implementation effort ≤ EUR 10k
- Action: Start after quick wins are in progress; use for low-priority FTE time

````python
fig, ax = plt.subplots(figsize=(12, 8))

colors_by_priority = {'HIGH': '#4dac26', 'MEDIUM': '#fee08b', 'LOW': '#fc8d59', 'DEFER': '#d73027'}
colors = [colors_by_priority.get(p, '#888') for p in opportunities['priority']]

scatter = ax.scatter(
    opportunities['impl_cost_eur'],
    opportunities['annual_saving_eur'],
    s=opportunities['frequency'] * 20,
    c=colors,
    alpha=0.7,
    edgecolors='black',
    linewidth=0.8,
)

for _, row in opportunities.iterrows():
    ax.annotate(row['workflow'][:20], (row['impl_cost_eur'], row['annual_saving_eur']),
                textcoords='offset points', xytext=(5, 3), fontsize=7)

median_cost = opportunities['impl_cost_eur'].median()
median_saving = opportunities['annual_saving_eur'].median()
ax.axvline(median_cost, color='gray', linestyle='--', alpha=0.5, linewidth=1)
ax.axhline(median_saving, color='gray', linestyle='--', alpha=0.5, linewidth=1)

ax.set_xlabel('Implementation Cost (EUR)', fontsize=11)
ax.set_ylabel('Annual Saving (EUR)', fontsize=11)
ax.set_title('Opportunity Quadrant: Implementation Effort vs Value\n(bubble size = frequency)', fontsize=12)
ax.grid(alpha=0.3)

legend_elements = [Patch(fc=c, label=p, edgecolor='black', linewidth=0.5) for p, c in colors_by_priority.items()]
ax.legend(handles=legend_elements, loc='upper left', fontsize=9)

plt.tight_layout()
plt.savefig('opportunity_quadrant.png', dpi=100, bbox_inches='tight')
plt.show()
````

## Part 6 — 100-Day Implementation Waves

**Wave 1 — Days 1–30: Quick Wins** (establish credibility)
- Objective: Deliver high value with minimal friction
- Criteria: High score + low effort + fast payback

**Wave 2 — Days 31–60: Foundation Projects** (build the platform)
- Objective: Deploy AI infrastructure for scaled automation
- Criteria: Medium score + medium effort + shared platform benefit

**Wave 3 — Days 61–90: Consolidation** (scale existing pilots)
- Objective: Expand Wave 2 pilots to other departments
- Criteria: Proven playbook + higher-effort migrations

**Wave 4 — Days 91+: Strategic** (transform the business)
- Objective: Execute longer-term, higher-risk projects
- Criteria: Score ≥ 15 + payback ≤ 18 months + strategic importance

````python
def assign_wave(row):
    '''Assign opportunity to wave 1-4 based on score and effort.'''
    score = row['score']
    effort = row['impl_cost_eur']
    payback = row['payback_months']
    if score >= 20 and effort <= 10000 and payback <= 12:
        return 1
    elif score >= 15 and effort <= 20000 and payback <= 15:
        return 2
    elif score >= 10 and effort <= 25000:
        return 3
    elif score >= 5:
        return 4
    else:
        return None

opportunities['wave'] = opportunities.apply(assign_wave, axis=1)

print('100-DAY IMPLEMENTATION WAVE PLAN')
print('=' * 70)
for wave in [1, 2, 3, 4]:
    if wave in opportunities['wave'].values:
        opps_in_wave = opportunities[opportunities['wave'] == wave]
        total_save = opps_in_wave['annual_saving_eur'].sum()
        print(f'Wave {wave}: {len(opps_in_wave)} opportunities, EUR {total_save:,.0f}/year')
        print(opps_in_wave[['workflow', 'score', 'annual_saving_eur', 'impl_cost_eur', 'payback_months']].to_string(index=False))
        print()
````

````python
wave_golive = {1: 30, 2: 60, 3: 90, 4: 120}

opps_with_wave = opportunities[opportunities['wave'].notna()].copy()
opps_with_wave['golive_day'] = opps_with_wave['wave'].map(wave_golive)

months = range(0, 13)
cumulative_saving = []

for month in months:
    day = month * 30
    saving = opps_with_wave[opps_with_wave['golive_day'] <= day]['annual_saving_eur'].sum() / 12 * month
    cumulative_saving.append(saving)

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(months, cumulative_saving, marker='o', color='#4dac26', linewidth=2.5, markersize=6)
ax.fill_between(months, cumulative_saving, alpha=0.3, color='#4dac26')
ax.set_xlabel('Months after Wave 1 go-live')
ax.set_ylabel('Cumulative Annual Saving (EUR)')
ax.set_title('100-Day Plan: Cumulative Value Build-Up')
ax.grid(alpha=0.3)
for m, s in zip(months, cumulative_saving):
    if m % 3 == 0:
        ax.text(m, s + 1000, f'EUR {s:,.0f}', ha='center', fontsize=9)
plt.tight_layout()
plt.savefig('cumulative_value.png', dpi=100, bbox_inches='tight')
plt.show()
````

## Part 7 — PE Investment Committee Deliverable Format

### What the IC Wants to Hear

1. **Executive summary (3 bullet points):** Total value, quick-win count, payback
2. **Top-10 opportunities table:** All 14 recommended fields from the scorecard
3. **100-day wave plan:** Wave-by-wave breakdown with go-live dates and saving
4. **Risk register:** Top 3 risks and mitigation strategies
5. **Competitive advantage:** How does this AI plan differentiate the portfolio company?

### What Gets Rejected

❌ **'We will replace the CFO with AI'** → No. You'll automate 5 workflows CFO currently owns.

❌ **'Estimated 50% margin improvement'** → Where's the detail? Show the 15 workflows.

❌ **'LLMs will solve this'** → Not without human control points. State the review cadence.

❌ **'Payback is infinite'** → If you can't quantify it, it's not on the scoring list.

✅ **'Invoice triage: 1,050 invoices/quarter × 5 min = 262 hrs/year = EUR 28k saving. Payback: 5 months.'**

````python
pe_summary = opportunities.head(10).copy()
pe_summary['roi_pct'] = ((pe_summary['annual_saving_eur'] - pe_summary['impl_cost_eur']) / pe_summary['impl_cost_eur'] * 100).round(0).astype(int)

total_portfolio_value = opportunities['annual_saving_eur'].sum()
total_implementation = opportunities['impl_cost_eur'].sum()
total_roi = (total_portfolio_value - total_implementation) / total_implementation * 100
quick_wins_count = len(opportunities[(opportunities['priority'] == 'HIGH') & (opportunities['payback_months'] <= 12)])

print('PE INVESTMENT COMMITTEE SUMMARY')
print('=' * 90)
print()
print('EXECUTIVE SUMMARY:')
print(f'- Total AI opportunity portfolio: EUR {total_portfolio_value:,.0f}/year, {total_roi:.0f}% 12-month ROI')
print(f'- Quick wins identified: {quick_wins_count} opportunities, <12-month payback')
print(f'- 100-day implementation plan: Wave 1 quick wins → Wave 2 foundation → Wave 3 scale → Wave 4 strategic')
print()
print('TOP 10 OPPORTUNITIES:')
print(pe_summary[['workflow', 'source_lesson', 'annual_saving_eur', 'impl_cost_eur', 'payback_months', 'roi_pct', 'priority']].to_string(index=False))
````

## Exercise 2 — Open-Ended: Defend Your Ranking
