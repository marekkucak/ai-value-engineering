## Module 00B: Value-Thinking Mindset Shift

From automation to value creation.

```python
def annual_manual_cost(volume, minutes, cost_per_hour):
    return volume * (minutes / 60) * cost_per_hour

def expected_run_rate_value(baseline, reduction, adoption, confidence):
    return baseline * reduction * adoption * confidence

print("✓ Value functions defined")
```

```python
workflows = pd.DataFrame({
    'workflow': ['support_triage', 'invoice_approval', 'access_review', 'saas_rationalization', 'knowledge_search'],
    'volume': [18000, 4200, 2400, 1800, 6000],
    'minutes': [14, 12, 8, 20, 5],
    'reduction': [0.45, 0.50, 0.55, 0.60, 0.65],
    'adoption': [0.80, 0.75, 0.90, 0.70, 0.85],
    'confidence': [0.90, 0.85, 0.95, 0.80, 0.90]
})

workflows['baseline'] = workflows.apply(lambda r: annual_manual_cost(r['volume'], r['minutes'], 55), axis=1)
workflows['expected_value'] = workflows.apply(
    lambda r: expected_run_rate_value(r['baseline'], r['reduction'], r['adoption'], r['confidence']), axis=1
)

print(workflows[['workflow', 'baseline', 'expected_value']].to_string())
print(f"\nTotal expected value: €{workflows['expected_value'].sum():,.2f}")
```

## Module 00C: Finance Crash Course

P&L structure and value mapping.

### P&L Structure
- Revenue
- COGS
- Operating Expenses (OpEx)
- EBITDA (key PE metric)
- Enterprise Value = EBITDA × Multiple (typically 8x)

```python
pnl_map = pd.DataFrame({
    'workflow': ['support_triage', 'invoice_approval', 'access_review', 'saas_rationalization', 'knowledge_search'],
    'p&l_line': ['OpEx', 'OpEx+Working Capital', 'OpEx', 'OpEx+Revenue', 'Revenue+OpEx'],
    'annual_value': [97192, 18313, 12234, 22680, 13851]
})

print("P&L Impact Mapping:")
print(pnl_map.to_string(index=False))
print(f"\nTotal: €{pnl_map['annual_value'].sum():,.2f}")
```

## Module 00D: Evidence→Cost→Impact Framework

Complete mini-case: Support Triage Automation

```python
print("="*70)
print("MINI-CASE: Support Triage Automation")
print("="*70)

volume = 18000
minutes = 14
cost_per_hour = 55
reduction = 0.45
adoption = 0.80
confidence = 0.90
implementation = 90000
recurring = 15000
ebitda_multiple = 8

baseline = annual_manual_cost(volume, minutes, cost_per_hour)
expected = expected_run_rate_value(baseline, reduction, adoption, confidence)
net_annual = expected - recurring
payback = implementation / net_annual
roi_yr1 = (net_annual - implementation) / implementation * 100
ev_impact = expected * ebitda_multiple

print(f"\nBASELINE:")
print(f"  {volume:,} cases × {minutes} min × €{cost_per_hour}/hr = €{baseline:,.2f}")
print(f"\nEXPECTED VALUE:")
print(f"  €{baseline:,.2f} × {reduction*100:.0f}% × {adoption*100:.0f}% × {confidence*100:.0f}% = €{expected:,.2f}")
print(f"\nNET ANNUAL VALUE:")
print(f"  €{expected:,.2f} - €{recurring:,} recurring = €{net_annual:,.2f}")
print(f"\nFINANCIAL METRICS:")
print(f"  Implementation cost: €{implementation:,}")
print(f"  Payback period: {payback:.1f} years")
print(f"  ROI (Year 1): {roi_yr1:.0f}%")
print(f"\nENTERPRISE VALUE IMPACT:")
print(f"  €{expected:,.2f} EBITDA × {ebitda_multiple}x = €{ev_impact:,.2f}")
print(f"\nRECOMMENDATION: {'✓ PROCEED' if payback < 2 else '⚠ REVIEW'}")
print("="*70)
```
